```python
#Sedat Öztürk
#E235013168

liste1 = []  

def odev1_veritopla():
    enazsayi = 5;
    sayisira = 1;
    
    #Burada ilk 5 sayı alınıyor.
    while len(liste1) < enazsayi:
        try:
            sayi = float(input(f"{sayisira}. sayıyı giriniz:"))
            
            if sayi in liste1: 
                print("Aynı sayı kullanılamaz. Lütfen farklı bir sayı giriniz.")
            else:
                liste1.append(sayi)
                sayisira += 1
        except ValueError:
            print("Lütfen geçerli bir sayı giriniz.")

    #Hayır diyene kadar başka sayı girmek istiyormusunuz sorusu soruluyor.
    while True:
        tekrar = input("Başka giriş yapacak mısınız? (Evet/E, Hayır/H): ").strip().lower()
        if tekrar in ['evet', 'e']:
            try:
                sayi = float(input(f"{sayisira}. sayıyı giriniz: "))
                
                if sayi in liste1: 
                    print("Aynı sayı kullanılamaz. Lütfen farklı bir sayı giriniz.")
                else:
                    liste1.append(sayi)
                    sayisira += 1
            except ValueError:
                print("Lütfen geçerli bir sayı giriniz.")
        elif tekrar in ['hayır', 'hayir', 'h']:
            break
        else:
            print("Lütfen 'Evet', 'E', 'Hayır' veya 'H' seçeneklerinden birini giriniz. (Dikkat!! Büyük veya küçük harf ayrımı yapılmaz.)")

    print("\nKaydedilen Sayılar:", liste1)

    pozitif = [sayi for sayi in liste1 if sayi > 0] #Pozitifler
    negatif = [sayi for sayi in liste1 if sayi < 0] #Negatifler
    sifir = [sayi for sayi in liste1 if sayi == 0] #Sıfır ama zaten varsa da 1 tane olabilir.

    print(f"\nGirilen sayıların {len(pozitif)} adedi pozitif, pozitif sayıların toplamı {sum(pozitif)}")
    print(f"Girilen sayıların {len(negatif)} adedi negatif, negatif sayıların toplamı {sum(negatif)}")
    print(f"Girilen sayıların {len(sifir)} adedi sıfır")

odev1_veritopla()
```

```python
def odev2_ortalama():
    liste2 = liste1.copy()
    ortalamadeger = sum(liste2) / len(liste2)
    
    silinenadet = 0 
    for sayi in liste2[:]:
        if sayi < ortalamadeger:
            liste2.remove(sayi)
            silinenadet += 1
            
    print(f"Liste1'in ortalaması {ortalamadeger}, ortalamadan küçük {silinenadet} adet sayı silindi ve Liste2 oluşturuldu")
    print("Liste1:", liste1)
    print("Liste2:", liste2)
    
odev2_ortalama()    
```

```python
def odev3_katilimcilistesi():
    katilimcilistesi = [
        "Mustafa Arat", "Sedat Öztürk", "Reha Erdağ", "Gazi Becit", "Karani İnce", 
        "Mehmet Albayrak", "Kürşat Gönüller", "Yavuz Acar", "Alper Ünal", "Remzi Baltacı"
    ]
    
    evetler = [] 
    
    for katilimci in katilimcilistesi:
        while True: #evet veya hayır dışında bir cevap verirse doğru cevabı verene kadar sorulacak.
            katilim_durumu = input(f"{katilimci} toplantıya katılacak mı? (Evet/E, Hayır/H): ").strip().lower()

            if katilim_durumu in ['evet', 'e']:
                evetler.append(katilimci) 
                break  
            elif katilim_durumu in ['hayır', 'hayir', 'h']:
                break  
            else:
                print("Lütfen 'Evet', 'E', 'Hayır' veya 'H' seçeneklerinden birini giriniz. (Dikkat!! Büyük veya küçük harf ayrımı yapılmaz.)")

    #katılımcılar dosyaya karışık şekilde yazılıyor.
    with open("KatilimciListesi.txt", "w") as karisik_dosya:
        for katilimci in evetler:
            karisik_dosya.write(katilimci + "\n")

    #katilimci listesi dosyası açılıp içerisindeki bilgiler listeye doldurulup sıralanacak 
    katilimcilar = []

    with open("KatilimciListesi.txt", "r") as dosya:
        for katilimci in dosya:
            katilimcilar.append(katilimci.strip()) 

    katilimcilar.sort()

    #katılımcılar dosyaya sıralanmış şekilde tekrar yazılıyor.
    with open("KatilimciListesi.txt", "w") as siralanmis_dosya:
        for katilimci in katilimcilar:
            siralanmis_dosya.write(katilimci + "\n")

    print(f"\nToplantıya {len(katilimcilar)} adet kişi katılacaktır, katılımcı listesi aşağıdadır:")
    
    with open("KatilimciListesi.txt", "r") as dosya:
        for satir in dosya:
            print(satir.strip()) 

odev3_katilimcilistesi()
```
